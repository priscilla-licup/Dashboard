## Visualization A: Key Metrics Cards

```python
def analyze_data_for_year_and_region(df, year, target_region):
    # Filter the DataFrame for the target region
    df_region = df[df['Region'] == target_region]

    # Check if the region exists in the DataFrame
    if df_region.empty:
        print(f"No data found for {target_region} in {year}.")
        return

    # Summarize total hazardous wastes
    total_hazardous_wastes = df_region['Total Hazardous Wastes'].sum()

    # Count waste disposal facilities by adding relevant columns
    total_disposal_facilities = df_region[['Illegal Dumpsites', 'Materials Recovery Facility', 'Sanitary Landfill', 'Registered TSD Facilities']].sum().sum()

    # Summarize population
    total_population = df_region['Population'].sum()

    return total_hazardous_wastes, total_disposal_facilities, total_population

# Specify the year and region you're interested in
target_year = 2020
target_region = 'Philippines'

# Assuming your datasets_by_year dictionary is defined as before
if target_year in datasets_by_year:
    df_year = datasets_by_year[target_year]
    hazardous_wastes, disposal_facilities, population = analyze_data_for_year_and_region(df_year, target_year, target_region)

    print(f"Data for {target_region} in {target_year}:")
    print(f"Total Hazardous Wastes: {hazardous_wastes}")
    print(f"Total Disposal Facilities: {disposal_facilities}")
    print(f"Population: {population}")
else:
    print(f"No data available for {target_year}.")
```

## Visualization B: Line Graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# Specific region to focus on
target_region = 'Philippines'

# Initialize a list to hold year and total hazardous waste data for the target region
yearly_data_for_region = []

# Collect data for the target region from each year
for year, df in datasets_by_year.items():
    total_waste = df.loc[df['Region'] == target_region, 'Total Hazardous Wastes'].sum()
    yearly_data_for_region.append((year, total_waste))

# Convert the list to a DataFrame for easier plotting
region_df = pd.DataFrame(yearly_data_for_region, columns=['Year', 'Total Hazardous Waste'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(region_df['Year'], region_df['Total Hazardous Waste'], marker='o', linestyle='-', color='b', label=target_region)

plt.title(f'Total Hazardous Waste Generated in {target_region} (Yearly)')
plt.xlabel('Year')
plt.ylabel('Total Hazardous Waste')
plt.legend()
plt.grid(True)
plt.xticks(region_df['Year'])  # Ensure all years are shown as x-ticks
plt.tight_layout()
plt.show()
```
